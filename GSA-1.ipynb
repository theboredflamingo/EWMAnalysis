```python
gdf['Reason'] = gdf['Reason'].astype(str)
```

```python
def setlabel(reason):
    reason.lower()
    if any(keyword in reason for keyword in ["eco friendly", "natural", "environment", "eco-friendly", "biodegradable", "sustainable", "organic", "clay"]):
        return "positive"
    elif any(keyword in reason for keyword in ["colorful", "color", "big idols", "artistic", "colors", "better", "shine"]):
        return "negative"
    else:
        return "neutral"
    
gdf["Label"] = gdf["Reason"].apply(setlabel)
        
        
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
def preprocess_text(text):
    
    tokens = word_tokenize(text.lower())
    filtered_tokens = [token for token in tokens if token not in stopwords.words('english')]
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]
    processed_text = ' '.join(lemmatized_tokens)

    return processed_text

# apply the function df

gdf['Reason'] = gdf['Reason'].apply(preprocess_text)
gdf
```

```python
nltk.download('vader_lexicon')
```

```python
analyzer = SentimentIntensityAnalyzer()
def get_sentiment(text):
    
    scores = analyzer.polarity_scores(text)
    sentiment = 1 if scores['pos'] > 0 else 0

    return sentiment

gdf['sentiment'] = gdf['Reason'].apply(get_sentiment)

gdf
```

```python
gdf["sentiment"].value_counts()
```

```python
def sentiment_to_int(sentiment):
    mapping = {'positive': 1, 'neutral': 0, 'negative': -1}
    return mapping.get(sentiment, 0)

gdf['Label'] = gdf['Label'].apply(sentiment_to_int)
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(gdf['Label'], gdf['sentiment']))
```
